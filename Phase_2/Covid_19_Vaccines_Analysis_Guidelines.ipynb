```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the Covid-19 vaccine data
data = pd.read_csv("/content/country_vaccinations.csv")
```

```python
# Basic statistics
summary = data.describe()
print(summary)
```

```python
# Plot vaccine coverage over time
fig, ax = plt.subplots()
data.plot(x='date', y='total_vaccinations', ax=ax)
plt.title('Covid-19 Vaccine Coverage Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.show()
```

```python
# Create an interactive dashboard for data exploration
fig = px.scatter(data, x='total_vaccinations', y='daily_vaccinations',
                 hover_data=['total_vaccinations_per_hundred'])
fig.update_layout(
    title="Covid-19 Vaccine Analysis Dashboard",
    xaxis_title="Total Vaccinations",
    yaxis_title="Infection Rate",
)
fig.show()
```
